```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    print(cmap)
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

```python
reeses = cv2.imread('DATA/reeses_puffs.png',0)   
display(reeses)
```

```python
cereals = cv2.imread('DATA/many_cereals.jpg',0)
display(cereals)
```

```python
# Create SIFT detector
sift = cv2.xfeatures2d.SIFT_create()
```

```python
# Detect keypoints and compute descriptors
kp1, des1 = sift.detectAndCompute(reeses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
```

```python
# Match features using BFMatcher with default params
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1, des2, k=2)
```

```python
# Apply ratio test to filter good matches
good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)
```

```python
 # Get the matching keypoints
src_pts= np.float32([kp1[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
dst_pts = np.float32([kp2[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)
```

```python
result = reeses.copy()
for i in src_pts:
    pt = tuple(i[0])
    cv2.circle(result,pt,10,(10),-1)
display(result)
```

```python
result = cereals.copy()
for i in dst_pts:
    pt = tuple(i[0])
    cv2.circle(result,pt,10,(10),-1)
display(result)
```

```python
#This calculates the homography matrix M, which defines a transformation from the object image space to the scene image space.
M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)
```

```python
M
```

```python
#Define the object image corners
h, w = reeses.shape
pts = np.float32([[0,0], [0,h], [w,h], [w,0]]).reshape(-1,1,2)
```

```python
pts
```

```python
#Transform corners using homography
dst = cv2.perspectiveTransform(pts, M)
```

```python
dst
```

```python
# Draw the rectangle on the scene
cereal_Finded = cv2.imread('DATA/many_cereals.jpg')
cereal_Finded = cv2.cvtColor(cereal_Finded,cv2.COLOR_BGR2RGB)
cv2.polylines(cereal_Finded, [np.int32(dst)], True, (0,255,0), 5, cv2.LINE_AA)
```
